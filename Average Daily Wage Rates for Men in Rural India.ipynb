```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("C:/Users/fatem/Downloads/Average Daily Wage.XLSX", skiprows=4 , usecols= "B:K")
df.head()
```

```python
df = df.dropna(subset=['State/Union Territory'])
df.head()
```

```python
desc_stats = df.describe()
round(desc_stats)
```

## Trend Analysis

```python
df.set_index ('State/Union Territory').T.plot(figsize=(12,8), legend= False)
plt.title('State-wise Average Daily Wage Rates Over the Years')
plt.xlabel('Year')
plt.ylabel('Wage Rate (₹)')
plt.legend(title='State/Union Territory', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
comparison_2022_23 = df[['State/Union Territory', '2022-23']].sort_values(by='2022-23', ascending=False)
comparison_2022_23
```

 Kerala stands out with a significantly higher average daily wage rate compared to other states, indicating a higher pay scale for construction workers in this region. The All India average wage rate is ₹393.3, which means states like Punjab, Rajasthan, and others listed before the All India average have higher than average wage rates. States like Bihar, West Bengal, Odisha, Gujarat, Tripura, and Madhya Pradesh have wage rates below the All India average, indicating relatively lower wages in these regions.


```python
df_growth = df.set_index('State/Union Territory').pct_change(axis='columns') * 100
df_growth = df_growth.reset_index()
df_growth
```

The data shows that states like Andhra Pradesh and Assam have seen steady and consistent growth in average daily wage rates over the years, suggesting stable economic conditions. In contrast, Meghalaya has experienced significant ups and downs, with wages dropping sharply in some years and rising dramatically in others, indicating economic instability or other influencing factors. Overall, there are notable differences in wage growth across regions, which can be attributed to various factors such as local economic conditions, state policies, and the demand for construction labor.
